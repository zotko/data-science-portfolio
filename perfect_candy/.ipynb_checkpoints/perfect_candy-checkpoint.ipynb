# Perfect Candy

## 1. Introduction

One famous candy company wants to create a brand new candy based on collected data on products in the market. Our task is to find product characteristics which drive customer sentiment and make a recommendation on a new candy.

For this case study we will use the [data set](https://github.com/fivethirtyeight/data/tree/master/candy-power-ranking) from [FiveThirtyEight](https://fivethirtyeight.com) that contains different information about different kinds of candy.

## 2. Importing modules and reading in the dataset

Let's start by reading the candies data to get a sense of different candy characteristics.

```python
# Import and show data
candies = pd.read_csv("candy-data.csv")
candies.head()
```

Let's give more understandable names to columns.

```python
# Rename columns
candies_update = candies.rename(
    {
        "competitorname": "Product",
        "chocolate": "Chocolate",
        "fruity": "Fruit Flavored",
        "caramel": "Caramel",
        "peanutyalmondy": "Peanuts/Almonds",
        "nougat": "Nougat",
        "crispedricewafer": "Crisped Rice/Wafers",
        "hard": "Hard",
        "bar": "Bar",
        "pluribus": "Bag/Box",
        "sugarpercent": "Sugar Percentile",
        "pricepercent": "Price Percentile",
        "winpercent": "Win Percentage",
    },
    axis=1,
)
```

## 4. Data pre-processing

For better visualization, we create another column "Soft" that contains the information whether the product is soft or not.

```python
candies_update["Soft"] = (candies_update["Hard"] == 0).replace({True: 1, False: 0})
```

Let's check if "Bar" and "Bag/Box" packaging types cover all products in the data set.

```python
candies_update[candies_update[["Bar", "Bag/Box"]].sum(axis=1) == 0].head()
```

We can see that some products have neither the 'Bar' nor the 'Bag/Box'  packaging. Let's create the column "Other Packaging" for those products.

```python
candies_update["Other Packaging"] = (
    candies_update[["Bar", "Bag/Box"]].sum(axis=1) == 0
).replace({True: 1, False: 0})
```

We'll drop rows with "One dime" and "One quarter" products as those are coins and not candies.

```python
candies_update = candies_update[
    ~candies_update.Product.isin(["One dime", "One quarter"])
]
```

Let's check if ingredient characteristics ('Chocolate', 'Fruit Flavored', 'Caramel', 'Peanuts/Almonds', 'Nougat', 'Crisped Rice/Wafers') cover all candies in the data set.

```python
candies_update[
    [
        "Product",
        "Chocolate",
        "Fruit Flavored",
        "Caramel",
        "Peanuts/Almonds",
        "Nougat",
        "Crisped Rice/Wafers",
    ]
][candies_update.loc[:, "Chocolate":"Crisped Rice/Wafers"].sum(axis=1) == 0]
```

We'll create a new column "Other Ingredients"  for candies which are not covered by ingredient variables.

```python
candies_update["Other Ingredients"] = (
    candies_update.loc[:, "Chocolate":"Crisped Rice/Wafers"].sum(axis=1) == 0
).replace({True: 1, False: 0})
```

```python
# reorder columns
candies_update = candies_update[
    [
        "Product",
        "Chocolate",
        "Fruit Flavored",
        "Caramel",
        "Peanuts/Almonds",
        "Nougat",
        "Crisped Rice/Wafers",
        "Other Ingredients",
        "Hard",
        "Soft",
        "Bar",
        "Bag/Box",
        "Other Packaging",
        "Sugar Percentile",
        "Price Percentile",
        "Win Percentage",
    ]
]
candies_update.head()
```

### 5.5 Relationship between All Variables

To get a better understanding of the relationships between all variables we'll compute the covariance matrix and visualize it with the help of heatmap.

```python
# Compute the correlation matrix
corr = candies_update.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
fig, ax = plt.subplots(figsize=(11, 15))


# Draw the heatmap with the mask and correct aspect ratio
with sns.axes_style("white"):
    _ = sns.heatmap(
        corr,
        mask=mask,
        linewidths=0.5,
        cmap="coolwarm",
        vmax=1,
        vmin=-1,
        annot=True,
        annot_kws={"size": 9},
        cbar_kws={"shrink": 0.4, "ticks": [-1, -0.5, 0, 0.5, 1]},
        square=True,
    )

# Add title
_ = plt.title("Pearson Correlation between All Variables", y=1.04)
```

With the of the heatmap, we can identify a positive correlation between win percentage and the presence of chocolate, peanuts/almonds, and a bar candy shape. There is also a strong negative correlation between chocolate and fruit flavors.

### 5.6 Relationship between Win Percentage and other Variables

Let's use mosaic plots to visualize the relationships between win percentage and other variables.

```python
def mosaic_plot(df, vars_x, vars_y, title=""):
    """Creates a mosaic plot."""
    data = (
        df[vars_y + vars_x]
        .melt(id_vars=vars_y)
        .query("value==1")
        .drop("value", axis=1)
        .set_index("variable")
        .stack()
    )

    data = data[data == 1].index.to_list()
    c = Counter(data)
    fig, ax = plt.subplots(figsize=(8, 8))
    colors = dict(zip(sorted(vars_y), sns.color_palette("Set2", 7)))

    _ = mosaic(
        c,
        ax=ax,
        gap=0.01,
        labelizer=lambda x: x[1] if x in c else "",
        properties={i: {"color": colors[i[1]]} for i in c},
        title=title,
    )

    legend = [(plt.Rectangle((0, 0), 1, 1, color=colors[i]), i) for i in sorted(vars_y)]
    _ = plt.yticks([], [])
```

First, we'll group numerical data into three bins: low (the lower quartile), medium (the interquartile range) and high (the upper quartile). Second, we'll use one-hot encoding to create a binary column for each category.

```python
for col in ["Win Percentage", "Price Percentile", "Sugar Percentile"]:
    data = pd.qcut(
        candies_update[col], q=[0.0, 0.25, 0.75, 1], labels=["Low", "Medium", "High"]
    )
    candies_update = candies_update.join(
        pd.get_dummies(data).add_suffix(f" {col.split()[0]}")
    )

candies_update.head()
```

```python
mosaic_plot(
    candies_update,
    ["Low Win", "Medium Win", "High Win"],
    ["Low Price", "Medium Price", "High Price"],
)

# Add title
_ = plt.title("Win Percentage vs. Price", y=1.04)
```

```python
mosaic_plot(
    candies_update,
    ["Low Win", "Medium Win", "High Win"],
    ["Low Sugar", "Medium Sugar", "High Sugar"],
)

# Add title
_ = plt.title("Win Percentage vs. Sugar Content", y=1.04)
```

More than half of top-ranked products have a medium price and a medium sugar content.

## 6. Feature Selection with Lasso Regression

Let's use the log transformation to obtain a bell-shaped distribution of our target variable - win percentage.

```python
log_win_percentage = np.log10(candies_update["Win Percentage"] + 1)

ax = sns.distplot(log_win_percentage, bins=20)

# Add labels and title
ax.set(xlabel="Log of Win Percentage", ylabel="Frequency")
_ = plt.title("Distribution of Win Percentage after Log Transformation", y=1.04)
```

We'll use a lasso regressor to select the most important features of our data set. To find the optimal hyperparameter alpha for the lasso regressor we'll use a grid search cross-validation.

```python
# Setup the hyperparameter grid
param_grid = {"alpha": np.logspace(-5, 8, 15)}

# Instantiate a lasso regressor
lasso = Lasso(normalize=True)

# Instantiate the GridSearchCV object
lasso_cv = GridSearchCV(lasso, param_grid, cv=5, iid=True)

# Select features
X = candies_update.loc[:, "Chocolate":"Price Percentile"].drop(
    ["Other Ingredients", "Soft", "Other Packaging"], axis=1
)
y = log_win_percentage

# Fit it to the data
lasso_cv.fit(X, y)

lasso_coef = lasso_cv.best_estimator_.coef_
```

To visualize the importance of candy attributes we'll plot the coefficients of the lasso regressor.

```python
data = pd.DataFrame({"Attribute": X.columns, "Coefficient": lasso_coef}).sort_values(
    "Coefficient", ascending=False
)

# Add labels and title
_ = sns.barplot(x="Attribute", y="Coefficient", data=data)
_ = plt.xticks(rotation=90)
_ = plt.title("Coefficients of Lasso Regressor", y=1.04)
```

From the plot above we can see that most important characteristic of a successful candy is the presence of chocolate. The next important feature is the amount of sugar that of course can correlate with the presence of chocolate or caramel. Other popular ingredients are peanuts/almonds, crisped rice/wafers, and fruit flavors. Caramel and nougat are not important for product success at all. Hard consistency of candy has a negative influence on the popularity of the product.

We can recommend to create a fruit flavored chocolate bar candy which contains peanuts/almonds and crisped rice/wafers.
