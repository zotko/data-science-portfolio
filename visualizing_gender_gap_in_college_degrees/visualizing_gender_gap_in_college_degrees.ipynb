# Visualizing the gender gap in college degrees

In this project, we'll work with the data from the [Department of Education Statistics](http://nces.ed.gov/programs/digest/2013menu_tables.asp) that contains the information on the percentage of bachelor's degrees granted to women from 1970 to 2012. The dataset was cleaned by Randal Olson and is available on his personal [website](http://www.randalolson.com/wp-content/uploads/percent-bachelors-degrees-women-usa.csv).

```python
import pandas as pd
import matplotlib.pyplot as plt
import textwrap

%matplotlib inline
```

```python
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
women_degrees.head()
```

```python
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)

stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

fig = plt.figure(figsize=(16, 20))

for ind, sp in enumerate(range(0, 16, 3)):
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[ind]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[ind]], c=cb_orange, label='Men', linewidth=3)
    ax.spines['right'].set_visible(False)    
    ax.spines['left'].set_visible(False)
    ax.spines['top'].set_visible(False)    
    ax.spines['bottom'].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    title = '\n'.join(textwrap.wrap(stem_cats[ind], width=15))
    if not ind:
        title = 'STEM\n\n\n' + title
    ax.set_title(title)
    ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
    ax.set_yticks([0, 100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    
    if sp == 0:
        ax.text(2005, 13, 'Men')
        ax.text(2002, 83, 'Women')
    elif sp == 15:
        ax.text(2005, 90, 'Men')
        ax.text(2001, 7, 'Women')
        ax.tick_params(labelbottom=True)
        
for ind, sp in enumerate(range(1, 14, 3)):
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[ind]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[ind]], c=cb_orange, label='Men', linewidth=3)
    ax.spines['right'].set_visible(False)    
    ax.spines['left'].set_visible(False)
    ax.spines['top'].set_visible(False)    
    ax.spines['bottom'].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    title = '\n'.join(textwrap.wrap(lib_arts_cats[ind], width=15))
    if not ind:
        title = 'LIBERAL ARTS\n\n' + title
    ax.set_title(title)
    ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
    ax.set_yticks([0, 100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    
    if sp == 1:
        ax.text(2005, 80, 'Men')
        ax.text(2002, 17, 'Women')
    if sp == 13:
        ax.tick_params(labelbottom=True)

for ind, sp in enumerate(range(2, 18, 3)):
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'],  women_degrees[other_cats[ind]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100- women_degrees[other_cats[ind]], c=cb_orange, label='Men', linewidth=3)
    ax.spines['right'].set_visible(False)    
    ax.spines['left'].set_visible(False)
    ax.spines['top'].set_visible(False)    
    ax.spines['bottom'].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    title = '\n'.join(textwrap.wrap(other_cats[ind], width=15))
    if not ind:
        title = 'OTHER\n\n' + title
    ax.set_title(title)
    ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
    ax.set_yticks([0, 100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    
    if sp == 2:
        ax.text(2005, 90, 'Men')
        ax.text(2002, 5, 'Women')
    elif sp == 17:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 30, 'Women')
        ax.tick_params(labelbottom=True)

plt.subplots_adjust(wspace=.3,
                   hspace=.3)
plt.show()
```

As you can see from the plot women dominated in 2012 in the following majors: Health Professions (85% women), Public Administration (82%), Education (79%), and Psychology (77%). The number of women in most of the [STEM](https://en.wikipedia.org/wiki/Science,_technology,_engineering,_and_mathematics) professions like Math,   Statistics, and Physical Sciences has increased to 40-45%. Even better, a majority of Biology degrees in 2012 (58%) were earned by women.
